NumPy Exercises

Exercise 1

Create a NumPy array and display its shape, number of dimensions, size, and datatype.

```python

import numpy as np
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])

print("Array:\n", arr1)
print("Shape:", arr1.shape)
print("Dimensions:", arr1.ndim)
print("Size:", arr1.size)
print("Datatype:", arr1.dtype)
```

Exercise 2

Create the following arrays using NumPy:

a 3×3 matrix of zeros

a 2×4 matrix of ones

a 3×3 identity matrix

```python
zeros = np.zeros((3,3))
ones = np.ones((2,4))
identity = np.eye(3)

print("\nZeros matrix:\n", zeros)
print("\nOnes matrix:\n", ones)
print("\nIdentity matrix:\n", identity)
```

Exercise 3

Generate numbers from 1 to 12 using np.arange() and reshape them into a 3×4 matrix.

```python
arr2 = np.arange(1,13)
reshaped = arr2.reshape(3,4)

print("\nOriginal array:", arr2)
print("\nReshaped matrix:\n", reshaped)
```

Exercise 4

Perform vector operations on two arrays:

addition

element-wise multiplication

dot product

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print("\nAddition:", a + b)
print("Element-wise multiplication:", a * b)
print("Dot product:", np.dot(a,b))
```

Exercise 5

Create a 3×3 matrix and perform slicing operations:

extract the first row

extract the first column

extract the top-left 2×2 submatrix

```python
matrix = np.arange(1,10).reshape(3,3)

print("\nMatrix:\n", matrix)
print("First row:", matrix[0])
print("First column:", matrix[:,0])
print("Top-left 2x2:\n", matrix[0:2,0:2])
```

Exercise 6

Combine two matrices using:

horizontal stacking

vertical stacking

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

horizontal = np.hstack((A,B))
vertical = np.vstack((A,B))

print("\nHorizontal Stack:\n", horizontal)
print("\nVertical Stack:\n", vertical)
```
