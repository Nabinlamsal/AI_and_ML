Question 2

Each operation should be implemented using separate functions, and the program must include error handling using try-except to manage invalid or empty inputs. Proper docstrings should also be included for each function.

```python

# Unit Converter
def unit_converter():
    """
    Convert between length, weight and volume units.
    """

    print("Choose conversion type:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    choice = input("Enter choice: ")

    try:
        value = float(input("Enter value: "))
    except ValueError:
        print("Invalid input")
        return

    if choice == "1":
        print("1. meters → feet")
        print("2. feet → meters")
        option = input("Choose option: ")

        if option == "1":
            print(value, "meters =", value * 3.28084, "feet")
        elif option == "2":
            print(value, "feet =", value / 3.28084, "meters")

    elif choice == "2":
        print("1. kg → pounds")
        print("2. pounds → kg")
        option = input("Choose option: ")

        if option == "1":
            print(value, "kg =", value * 2.20462, "pounds")
        elif option == "2":
            print(value, "pounds =", value / 2.20462, "kg")

    elif choice == "3":
        print("1. liters → gallons")
        print("2. gallons → liters")
        option = input("Choose option: ")

        if option == "1":
            print(value, "L =", value * 0.264172, "gallons")
        elif option == "2":
            print(value, "gallons =", value / 0.264172, "liters")


# Mathematical operations on list
def find_sum(numbers):
    return sum(numbers)

def find_average(numbers):
    return sum(numbers) / len(numbers)

def find_max(numbers):
    return max(numbers)

def find_min(numbers):
    return min(numbers)


# Example run
nums = [10,20,30,40]

print("Sum:", find_sum(nums))
print("Average:", find_average(nums))
print("Max:", find_max(nums))
print("Min:", find_min(nums))
```

Question 9

Write a Python function that extracts elements from a list in reverse order starting from the second-to-last element and skipping one element each time. For example, given [1,2,3,4,5,6], the output should be [5,3,1].

```python
#List Manipulation
# Extract every other element
def extract_every_other(lst):
    return lst[::2]

# Slice sublist
def get_sublist(lst, start, end):
    return lst[start:end+1]

# Reverse list
def reverse_list(lst):
    return lst[::-1]

# Remove first and last element
def remove_first_last(lst):
    return lst[1:-1]

# Get first n elements
def get_first_n(lst, n):
    return lst[:n]

# Get last n elements
def get_last_n(lst, n):
    return lst[-n:]

# Reverse skip elements
def reverse_skip(lst):
    return lst[-2::-2]


# Example tests
lst = [1,2,3,4,5,6]

print("Every other:", extract_every_other(lst))
print("Sublist:", get_sublist(lst,2,4))
print("Reverse:", reverse_list(lst))
print("Remove first/last:", remove_first_last(lst))
print("First 3:", get_first_n(lst,3))
print("Last 2:", get_last_n(lst,2))
print("Reverse skip:", reverse_skip(lst))
```

Write a Python function that calculates the sum of all numbers in a nested list regardless of depth. For example, [[1,2],[3,[4,5]],6] should return 21.

Write a Python function that removes all occurrences of a specific element from a nested list. For example, removing 2 from [[1,2],[3,2],[4,5]] should produce [[1],[3],[4,5]].

Write a Python function that finds the maximum value in a nested list regardless of its depth.

Question 15

Write a Python function that counts how many times a specific element appears in a nested list.

Question 16

Write a Python function that deeply flattens a list of lists of lists into a single list containing all elements.

Question 17

Write a Python function that calculates the average of all numerical elements in a nested list.

```python

# Flatten nested list
def flatten(lst):
    result = []
    for i in lst:
        if isinstance(i,list):
            result.extend(flatten(i))
        else:
            result.append(i)
    return result


# Access nested element
def access_nested(lst, indices):
    return lst[indices[0]][indices[1]]


# Sum nested list
def sum_nested(lst):
    return sum(flatten(lst))


# Remove element from nested list
def remove_element(lst, elem):
    result=[]
    for i in lst:
        if isinstance(i,list):
            result.append(remove_element(i,elem))
        else:
            if i != elem:
                result.append(i)
    return result


# Find max element
def find_max(lst):
    return max(flatten(lst))


# Count occurrences
def count_occurrences(lst, elem):
    return flatten(lst).count(elem)


# Deep flatten
def deep_flatten(lst):
    return flatten(lst)


# Nested average
def average_nested(lst):
    flat = flatten(lst)
    return sum(flat)/len(flat)


# Example tests
nested = [[1,2],[3,[4,5]],6]

print("Flatten:", flatten(nested))
print("Sum:", sum_nested(nested))
print("Max:", find_max(nested))
print("Count of 2:", count_occurrences([[1,2],[2,3],[2,4]],2))
print("Average:", average_nested([[1,2],[3,4],[5,6]]))
```
